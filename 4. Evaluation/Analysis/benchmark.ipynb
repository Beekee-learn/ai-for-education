# Gsm8k

```python
import os
import json

# List all directories in the current repository except .venv, hidden folders, and folders containing "rag"
folders = [f for f in os.listdir('.') if os.path.isdir(f) and f != '.venv' and not f.startswith('.') and 'RAG' not in f]
print(folders)

import matplotlib.pyplot as plt

results = {}

for folder in folders:
    try:
        with open(os.path.join(folder, 'gsm8k', 'results.json')) as f:
            data = json.load(f)
            try:
                results[folder] = data["average"]["general_results"]
            except KeyError:
                print(f"Skipping folder {folder} due to missing key")
    except (FileNotFoundError, KeyError):
        print(f"Skipping folder {folder} due to missing file or key")

# Sort the results dictionary by general_results value
sorted_results = dict(sorted(results.items(), key=lambda item: item[1], reverse=True))

# Plotting the sorted results with Llama 2 in red
colors = ['orange' if 'gpt 3.5' in key else 'red' if 'Llama 2' in key else 'blue' for key in sorted_results.keys()]

plt.bar(sorted_results.keys(), sorted_results.values(), color=colors)
plt.xlabel('Models')
plt.ylabel('General Results')
plt.title('General Results on GSM8K dataset')
plt.xticks(rotation=90)
plt.show()
```
